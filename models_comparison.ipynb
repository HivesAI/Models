```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
data = pd.read_csv("models_performance.csv", index_col='Model') 
print(data)
```

```python
piante = list(data.columns[1:])
colors = ["blue", "green", "black", "cyan", "orange", "red", "orange", "purple"]

for pianta in piante[:1]:
    fig, ax = plt.subplots(figsize=(15, 6), dpi=80)
    ax.set(ylim=(0., 1.))
    for i in range(len(data.index)):
        model = data.index[i]
        ax.plot(piante, np.maximum(data.loc[model, piante], 0), color=colors[i], label=model)
    ax.legend()
```
